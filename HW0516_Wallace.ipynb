```python
#字詞數/句子數計算
from nltk.corpus import gutenberg
for fileid in gutenberg.fileids():

  num_chars=len(gutenberg.raw(fileid))#輸出文本原始內容
  num_words=len(gutenberg.words(fileid))#輸出文本單詞列表
  num_sents=len(gutenberg.sents(fileid))#輸出文本句子列表
  print(num_chars, num_words, num_sents, fileid)
```

```python
from nltk.corpus import gutenberg
for fileid in gutenberg.fileids():

  num_chars=len(gutenberg.raw(fileid))#輸出文本原始內容
  num_words=len(gutenberg.words(fileid))#輸出文本單詞列表
  num_sents=len(gutenberg.sents(fileid))#輸出文本句子列表
  num_vocab=len(set(w.lower()for w in gutenberg.words(fileid)))#將w轉換成小寫
  #計算平均字詞長度，平均句子長度，字彙多樣性
  print(round(num_chars/num_words), #round()四捨五入
     round(num_words/num_sents),
     num_vocab/num_words, fileid)
```

```python
#結合分類與特定字詞的應用(frequency distribution)
import nltk
from nltk.corpus import brown
#使用FreqDist()在單一文本計算特定字詞的個數
reviews_text= brown.words(categories='reviews')
fdist=nltk.FreqDist(w.lower()for w in reviews_text)
#fdist:{can:45, it:20, news:10}fdist{"can"}->45
modals=['can','could','may','might','must','will','shall']
#找助動詞的字數
for m in modals:
  print(m+':', fdist[m])
```

```python
from nltk.corpus import brown
#使用ConditionalFreqDist()在多個文本計算特定字詞的個數
#FreqDist()僅在單一文本計算特定字詞的個數
cfd= nltk.ConditionalFreqDist((genre,word)
  for genre in brown.categories()
  for word in brown.words(categories=genre))
genres=['reviews','hobbies','romance','humor']
modals=['can','could','may','might','must','will','shall','fuck']
#cfd.tabulate(conditions=genres, samples=modals)
cfd.tabulate(conditionals=genres, samples=modals)
#conditions僅列出你標記的， conditionals會跑出全部
#sample是橫項
```
