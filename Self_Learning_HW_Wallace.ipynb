```python
#Hamlet频率分布图

import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
import re

# 定义去除标点符号的函数
def remove_punctuation(input_str):
    return re.sub(r'[^\w\s]', '', input_str)

# 下载必要的资源
nltk.download('stopwords')
nltk.download('punkt')

# 读取文本文件内容
HW_1 = nltk.corpus.gutenberg.raw('shakespeare-hamlet.txt')

# 去除文本中的标点符号
HW_1_no_punct = remove_punctuation(HW_1)

# 将文本进行分词
tokens = word_tokenize(HW_1_no_punct)

# 获取停用词列表
stop_words = set(stopwords.words('english'))

# 移除停用词
clean_tokens = [token for token in tokens if token.lower() not in stop_words]

# 输出移除停用词后的词汇列表
print(clean_tokens)

# 计算词频
freq = nltk.FreqDist(clean_tokens)

# 打印词频项目
print(freq.items())

# 画出前20个词的频率分布图
freq.plot(20, cumulative=False)
```

```python
import nltk
from nltk.corpus import PlaintextCorpusReader
from nltk.probability import ConditionalFreqDist

# 下载必要的nltk数据包（如果还没有下载）
nltk.download('punkt')

# 指定文本所在目录
corpus_root = '/content/drive/MyDrive/Colab'

# 读取文本文件
wordlists = PlaintextCorpusReader(corpus_root, '.*\.txt')

# 获取文本文件列表
books = wordlists.fileids()

# 创建 ConditionalFreqDist 对象
cfd = ConditionalFreqDist(
    (fileid, word.lower())  # 这里将词汇转换为小写，以确保统计的一致性
    for fileid in books
    for word in wordlists.words(fileid)
)

# 指定要查找的词汇列表
findList = ["love", "hate", "fear", "joy", "sadness", "anger", "happy", "unhappy", "grief", "sorrow",
            "rage", "jealousy", "envy", "pity", "despair", "hope", "desire", "dread", "calm", "fury",
            "tears", "smile", "cry", "laugh", "pain", "suffer", "torment", "pleasure", "delight", "anguish"]

# 使用tabulate方法显示词频统计结果
cfd.tabulate(conditions=books, samples=findList)

print("---分隔線---")

# 打印文本文件列表
print(books)
```
